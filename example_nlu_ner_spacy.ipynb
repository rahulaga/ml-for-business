```python
import spacy
from spacy import displacy
import pandas as pd
```

```python
text = "Joe Biden’s top trade adviser Katherine Tai said that while the US administration 'believes strongly in IP protections, it would support a waiver of those rules for Covid-19 vaccines."
```

```python
# load the model, ref https://spacy.io/usage/models
nlp = spacy.load('en_core_web_sm')
doc = nlp(text)
```

```python
# NER = named entity recoginition
displacy.render(doc, style='ent', jupyter=True)
```

```python
doc.ents
```

```python
for ent in doc.ents:
  print(ent, ent.label_, ent.start_char, ent.end_char)
```
