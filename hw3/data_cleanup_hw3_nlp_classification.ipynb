```
#rename columns - remove spaces
data.rename(columns={'Review Text':'Review_Text', 'Recommended IND':'Recommended_IND','Division Name':'Division_Name','Department Name':'Department_Name','Class Name':'Class_Name'}, inplace=True)
```

Further cleanup for own model and also needed for Google

```
#remove new lines etc
data.replace(to_replace=[r"\\t|\\n|\\r", "\t|\n|\r"], value=["",""], regex=True, inplace=True)
```

```
#many rows have no reviews or title
data.isna().sum()
```

```
categorical_columns=['Division_Name','Department_Name','Class_Name']
#data[categorical_columns].value_counts()
```

```
# remove rows with no value for these columns (less than 50 rows)
data.dropna(subset=categorical_columns, inplace=True)
```

```
#remove rows where both Title and Review text are missing
data.dropna(how='all', subset=['Title','Review_Text'], inplace=True)
data.isna().sum()
```

```
#Fill empty title with blank
data['Title'].fillna('', inplace=True)
data.isna().sum()
```

```
#copy over title (if present) into Review text so then there is only 1 text field
data['Title_Review'] = data.Title.astype(str).str.cat(data['Review_Text'].astype(str), sep=' ')
data.head()
```

```
#remove title and review text
del data['Title']
del data['Review_Text']
```
